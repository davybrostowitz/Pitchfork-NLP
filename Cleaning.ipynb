```python
import pandas as pd
```

```python
data = pd.read_csv('pitchfork_reviews2.csv')
```

```python
data.head(10)
```

```python
data = data.dropna(axis = 0)
```

```python
data.head(10)
```

```python
dates = data['date'].str.split(' ',expand = True)
```

```python
dates.head(10)
```

```python
data['Month'] = dates[0]
data['Day'] = dates[1]
data['Year'] = dates[2]
```

```python
data.head(10)
```

```python
data.genre.unique()
```

```python
genres = data['genre'].str.split(',',expand = True)
```

```python
genres.head(10)
```

```python
data['Genre1'] = genres[0]
data['Genre2'] = genres[1]
data['Genre3'] = genres[2]
data['Genre4'] = genres[3]
```

```python
data.head(2)
```

```python
data = data.drop(['date','genre'], axis = 1)
```

```python
data.head(10)
```

```python
#preprocessing
import re

# Convert all the string to lower cases
data['content'] = data['content'].str.lower()
# \s+ means all empty space (\n, \r, \t)
data['content'] = data['content'].apply(lambda x: re.sub('\s+', ' ', x))
# We don't want empty string in our text
data = data.loc[data['content'] != ""]
```
